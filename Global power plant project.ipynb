# Overview
The Global Power Plant Database project aims to provide a comprehensive, open-source repository of power plant information from around the world. It centralizes data on approximately 14,000 power plants across three countries: the USA, Australia, and India. This database covers a wide range of power generation sources, including thermal plants (such as coal, gas, oil, nuclear, biomass, waste, and geothermal) as well as renewables (including hydro, wind, and solar). Each power plant entry contains key information such as its capacity, generation, ownership, geographical coordinates, and fuel types.

# Data description
The key attributes of the database include the following indicators: country (the ISO 3166-1 alpha-3 country code and its longer form), power plant name, unique identifier (gppd_idnr), electrical generating capacity in megawatts (capacity_mw), geolocation coordinates (latitude and longitude), primary fuel type (primary_fuel), additional fuel types if applicable (other_fuel1, other_fuel2, other_fuel3), commissioning year (commissioning_year), majority shareholder (owner), data source (source and url), geolocation attribution (geolocation_source), unique plant identifier (wepp_id), year of capacity data (year_of_capacity_data), electricity generation in gigawatt-hours reported for various years from 2013 to 2019 (generation_gwh_2013 to generation_gwh_2019), data source attribution for generation information (generation_data_source), estimated electricity generation in gigawatt-hours for the same years (estimated_generation_gwh_2013 to estimated_generation_gwh_2017), and the model/method label used for estimation (estimated_generation_note_2013 to estimated_generation_note_2017)

# Model Selection and Training
A variety of machine learning models were explored, such as logistic regression, k-neighbors classifier, decision tree regressor, decsion tree classifier, support vector classifier etc. These models were selected for their capacity to handle regression and classification tasks and their flexibility in capturing complex relationships within the data. Prior to training, the dataset was partitioned into training and testing subsets using the train_test_split function from the scikit-learn library, ensuring robust model evaluation and performance assessment.

# Hyperparameter Tuning
Hyperparameter tuning was performed for some models to optimize their performance. This process involved selecting the best values for the hyperparameters using techniques such as grid search or randomized search. Evaluation metrics were used to compare different models and select the best-performing one.
