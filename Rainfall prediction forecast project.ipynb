```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
rp['RainToday'].value_counts().plot.pie(autopct='%1.1f%%')
centre=plt.Circle((0,0),0.7,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre)
plt.subplot(1,2,2)
sns.countplot(x='RainToday',data=rp)
plt.xticks(rotation=45)
rp['RainToday'].value_counts()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
rp['RainTomorrow'].value_counts().plot.pie(autopct='%1.1f%%')
centre=plt.Circle((0,0),0.7,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre)
plt.subplot(1,2,2)
sns.countplot(x='RainTomorrow',data=rp)
plt.xticks(rotation=45)
rp['RainTomorrow'].value_counts()
```

```python
sns.pairplot( data=rp, vars=('MaxTemp','MinTemp','Pressure9am','Pressure3pm'), hue='RainTomorrow' )
plt.show()
```
